```python
from dataloader import tensor_to_image, val_dataset, train_dataset, val_dataset
import numpy as np
import torchvision
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from natsort import natsorted
import os
```

```python
l = np.random.randint(0, len(train_dataset)-3)

# print(l)

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 10))

# cols = ["Image A", "Image B", "Label"]
# rows = ['Row {}'.format(row) for row in ['A', 'B', 'C']]
# for ax, col in zip(axes[0], cols):
#     ax.set_title(col)
# for ax, row in zip(axes[:,0], rows):
#     ax.set_ylabel(row, rotation=0, size='large')

fig.tight_layout()

for i in range(l, l+1):
    sample = train_dataset[i]
    
    # a = natsorted (os.listdir( os.path.join("./Datasets/images_split/train","A") ))
    # print(os.path.join("./Datasets/images_split/train", "A", a[i]))
    # a = natsorted (os.listdir( os.path.join("./Datasets/images_split/train","B") ))
    # print(os.path.join("./Datasets/images_split/train", "B", a[i]))
    # a = natsorted (os.listdir( os.path.join("./Datasets/images_split/train","label") ))
    # print(os.path.join("./Datasets/images_split/train", "label", a[i]))

    plt.subplot(1, 3, 3*(i-l) + 1)
    plt.imshow(tensor_to_image(sample["img_A"]))

    plt.subplot(1, 3, 3*(i-l) + 2)
    plt.imshow(tensor_to_image(sample["img_B"]))

    plt.subplot(1, 3, 3*(i-l) + 3)
    plt.imshow(tensor_to_image(sample["label"]), cmap="gray")    
```
