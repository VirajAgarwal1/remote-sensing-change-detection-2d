```python
input1 = np.zeros((3,256,256))
index1 = random.randint(0,255)
index2 = random.randint(0,255)
for j in range(3):
    input1[j][index1][index2] = 1
input1 = input1.astype("float32")
input1 = torch.Tensor(input1)

input2 = torch.Tensor(np.zeros((3,256,256)).astype("float32"))

input = torch.cat([input1.reshape((1,3,256,256)), input2.reshape((1,3,256,256))], dim=0)

model = RDPNet(3, 2)
out = model(input1.reshape((1,3,256,256)), input2.reshape((1,3,256,256)))
out = out.detach().numpy()
print(out.shape)
out_fix = out[0][0].copy()
for i in range(256):
    for j in range(256):
        out_fix[i][j] = np.argmax( [out[0][0][i][j], out[0][1][i][j]] )
out_fixout_fix = out_fix.reshape(256,256,1)
plt.imshow(out_fix, cmap="Greys")
```
