```python
from dataloader import tensor_to_image, val_dataset, train_dataset, val_dataset
import numpy as np
import matplotlib.pyplot as plt
```

```python
numPlots = 3
l = np.random.randint(0, len(train_dataset)-numPlots)

fig, axes = plt.subplots(nrows=numPlots, ncols=2, figsize=(10, 10))
fig.tight_layout()

labels = []

for i in range(l, l+numPlots):
    sample = train_dataset[i]
    
    plt.subplot(numPlots, 2, 2*(i-l) + 1)
    plt.imshow(tensor_to_image(sample["img_A"]))

    plt.subplot(numPlots, 2, 2*(i-l) + 2)
    plt.imshow(tensor_to_image(sample["img_B"]))

    labels.append(sample["label"])

plt.show()
for lbl in labels:
    print(lbl)
```
